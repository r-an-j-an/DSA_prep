```python
class Fraction:
    def __init__(self,a,b):
        self.a=a
        self.b=b
        if(self.b==0):
            print("Denominator Cannot be Zero. Please Check...")
            delattr(self,"a")
            delattr(self,"b")
            
    def fprint(self):
        try:
            if(self.b == 1):
                print(self.a)
            else:
                print(self.a," / ", self.b)
        except:
            print("Error!!..")
    def fadd(self,f2):
        try:
            num=(self.a*f2.b)+(f2.a*self.b)
            den=(self.b*f2.b)
            res=Fraction(num,den)
            print("The addition of ",self.a," / ",self.b ," and ",f2.a," / ",f2.b ," is ",res.a," / ",res.b )
            return res
        except:
            print("Error!!..")
            
    
    def fsub(self,f2):
        try:
            num=(self.a*f2.b)-(f2.a*self.b)
            den=(self.b*f2.b)
            res=Fraction(num,den)
            print("The difference of ",self.a," / ",self.b ," and ",f2.a," / ",f2.b ," is ",res.a," / ",res.b )
            return res
        except:
            print("Error!!..")
            
    def fmul(self,f2):
        try:
            res=Fraction(self.a*f2.a,self.b*f2.b)
            print("The multiplication of ",self.a," / ",self.b ," and ",f2.a," / ",f2.b ," is ",res.a," / ",res.b )
            return res
        except:
            print("Error!!..")
            
    def fsimpl(self):
        if(self.b == 1 or self.a == 0):
            return self
        if(self.b==self.a):
            self.a=1
            self.b=1
                
        for i in range(2,min(self.a,self.b)):
            if(self.a%i==0 and self.b%i==0):
                self.a=self.a//i
                self.b=self.b//i
                break
        return self
            
        
```

```python
f1=Fraction(2,3)
f1.fprint()
```

```python
f2=Fraction(10,10)
f2.fprint()
f2=f2.fsimpl()
f2.fprint()
```

```python
res=f1.fadd(f2)
res=res.fsimpl()
res.fprint()
```

```python
res=f1.fsub(f2)
res=res.fsimpl()
res.fprint()
```

```python
res=f1.fmul(f2)
res=res.fsimpl()
res.fprint()
```

```python
res=res.fsimpl()
res.fprint()
```
